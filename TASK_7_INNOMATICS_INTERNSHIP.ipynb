### QUESTION 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self):
        self.name = ""
        self.state = "OFF"
    def take_input(self):
        self.name = input("Enter Light Name: ")
        self.state = input("Enter Action (ON/OFF): ")
    def update_state(self):
        if self.state == "ON":
            self.state = "ON"
        else:
            self.state = "OFF"
    def show(self):
        print(self.name, "is", self.state)
light = SmartLight()
light.take_input()
light.update_state()
light.show()
```

### QUESTION 2: Employee ID Card System

```python
class EmployeeCard:
    def __init__(self):
        self.name = ""
        self.emp_id = ""
        self.dept = ""
    def collect_details(self):
        self.name = input("Enter Employee Name: ")
        self.emp_id = input("Enter Employee ID: ")
        self.dept = input("Enter Department: ")
    def print_card(self):
        print("\n Employee ID Card ")
        print("Name      :", self.name)
        print("ID        :", self.emp_id)
        print("Department:", self.dept)
card = EmployeeCard()
card.collect_details()
card.print_card()
```

### QUESTION 3: Mobile Contact Record

```python
class ContactBook:
    def __init__(self):
        self.name = ""
        self.phone = ""
    def add_contact(self):
        self.name = input("Enter Contact Name: ")
        while True:
            self.phone = input("Enter Phone Number: ")
            if len(self.phone) == 10:
                break
            else:
                print("Enter valid 10-digit number!")
    def display(self):
        print("\nContact Saved Successfully")
        print("Name :", self.name)
        print("Phone:", self.phone)
contact = ContactBook()
contact.add_contact()
contact.display()
```

### QUESTION 4: Product Price Tag Generator

```python
class PriceTag:
    def __init__(self):
        self.product = ""
        self.cost = 0
    def read_data(self):
        self.product = input("Enter Product Name: ")
        self.cost = int(input("Enter Price: "))
    def print_tag(self):
        print("\n PRICE TAG ")
        print("Product :", self.product)
        print("Price   : ₹", self.cost)
tag = PriceTag()
tag.read_data()
tag.print_tag()
```

### QUESTION 5: Movie Rating Display System

```python
class MovieReview:
    def __init__(self):
        self.movie = ""
        self.rating = 0
    def get_info(self):
        self.movie = input("Enter Movie Name: ")
        self.rating = float(input("Enter Rating: "))
    def show_review(self):
        if self.rating >= 4.5:
            level = "Excellent"
        elif self.rating >= 3.5:
            level = "Good"
        else:
            level = "Average"
        print("\nMovie:", self.movie)
        print("Rating:", self.rating, "/ 5")
        print("Review:", level)
review = MovieReview()
review.get_info()
review.show_review()
```
