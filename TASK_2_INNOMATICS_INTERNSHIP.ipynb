```python
#Question 1:UNIQUE WORDS IN A SENTENCE
```

```python
#METHOD-1
s = input("Enter sentence: ")

w = s.split()
u = set(w)

print("Unique words count:", len(u))
print("Unique words:", u)
```

```python
#METHOD-2
s = input("Enter sentence: ")

w = s.split()
u = []

for i in w:
    if i not in u:
        u.append(i)

print("Unique words count:", len(u))
print("Unique words:", u)
```

```python
#Question 2:HIGHEST SALARY FROM EMPLOYEE DATA
```

```python
#METHOD-1
n = int(input("Enter number of employees: "))
d = {}
for i in range(n):
    x = input("Enter name: ")
    e = int(input("Enter salary: "))
    d[x] = e
k = max(d, key=d.get)
print("Highest Salary:", k, "-", d[k])
```

```python
#METHOD-2
n = int(input("Enter number of employees: "))
d = {}
for i in range(n):
    x = input("Enter name: ")
    e = int(input("Enter salary: "))
    d[x] = e
m = 0
k = ""
for i in d:
    if d[i] > m:
        m = d[i]
        k = i
print("Highest Salary:", k, "-", m)
```

```python
#Question 3:FIND MAXIMUM AND MINIMUM VALUES
```

```python
#METHOD-1
l = list(map(int, input("Enter numbers: ").split()))
print(f"List:{l}")
print("Max:", max(l))
print("Min:", min(l))
```

```python
#METHOD-2
l = list(map(int, input("Enter numbers: ").split()))
mx = l[0]
mn = l[0]
for i in l:
    if i > mx:
        mx = i
    if i < mn:
        mn = i
print(f"List:{l}")
print("Max:", mx)
print("Min:", mn)
```

```python
#Question 4:COUNT PRODUCTS ABOVE A PRICE THRESHOLD
```

```python
#METHOD-1
l = list(map(int, input("Enter prices: ").split()))
c = sum(1 for i in l if i > 1000)
print("Products above 1000:", c)
```

```python
#METHOD-2
l = list(map(int, input("Enter prices: ").split()))
c = 0
for i in l:
    if i > 1000:
        c += 1
print("Products above 1000:", c)
```

```python
#Question 5:CALCULATE ATTENDANCE PERCENTAGE
```

```python
#METHOD-1
l = input("Enter attendance: ").split()
l = [i.upper() for i in l]
p = l.count("P")
per = (p / len(l)) * 100
print("Attendance Percentage:", per)
```

```python
#METHOD-2
l = input("Enter attendance: ").split()
l = [i.upper() for i in l]
p = 0
for i in l:
    if i == "P":
        p += 1
per = (p / len(l)) * 100
print("Attendance Percentage:", per)
```

```python
#Question 7:COUNT CHARACTER FREQUENCY
```

```python
#METHOD-1
s = input("Enter string: ")
d = {}
for i in s:
    d[i] = d.get(i, 0) + 1
print(d)
```

```python
#METHOD-2
s = input("Enter string: ")
d = {}
for i in s:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
```

```python
#Question 8:CONVERT LIST TO TUPLE
```

```python
#METHOD-1
l = list(map(int, input("Enter numbers: ").split()))
t = tuple(l)
print("Tuple:", t)
```

```python
#METHOD-2
l = list(map(int, input("Enter numbers: ").split()))
t = ()
for i in l:
    t = t + (i,)
print("Tuple:", t)
```

```python
#Question 9:CHECK IF A KEY EXISTS IN A DICTIONARY
```

```python
#METHOD-1
d = {"Ravi": 75000, "Anita": 68000}
x = input("Enter name to search: ").capitalize()
if x in d:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#METHOD-2
d = {"Ravi": 75000, "Anita": 68000}
x = input("Enter name to search: ").capitalize()
f = False
for i in d:
    if i == x:
        f = True
if f:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
#Question 10:CALCULATE AVERAGE MARKS
```

```python
#METHOD-1
l = list(map(int, input("Enter marks: ").split()))
a = sum(l) / len(l)
print("Average Marks:", a)
```

```python
#METHOD-2
l = list(map(int, input("Enter marks: ").split()))
s = 0
for i in l:
    s += i
a = s / len(l)
print("Average Marks:", a)
```
