### QUESTION 1: Social Media – Post Engagement Analyzer

```python
def post_check():
    data = list(map(int, input("Enter like counts: ").split()))
    total_likes = 0
    for val in data:
        total_likes += val
    print("Total Likes:", total_likes)
    status = "Normal Engagement"
    if total_likes >= 1000:
        status = "Viral Post"
    print("Post Status:", status)
post_check()
```

### QUESTION 3: Agriculture – Rainfall Adequacy Checker

```python
def rain_report():
    readings = list(map(int, input("Enter rainfall readings: ").split()))
    total_water = 0
    days = 0
    for r in readings:
        total_water += r
        days += 1
    avg_rain = total_water / days
    print("Average Rainfall:", avg_rain)
    result = "Inadequate Rainfall"
    if avg_rain >= 70:
        result = "Adequate Rainfall"
    print("Rainfall Status:", result)
rain_report()
```

### QUESTION 4: Social Media – Duplicate Account Detection

```python
def user_scan():
    names=input("Enter usernames(with spaces): ").split()
    names=[name.lower() for name in names]
    duplicate = False
    i = 0
    while i < len(names):
        j = i + 1
        while j < len(names):
            if names[i] == names[j]:
                duplicate = True
            j += 1
        i += 1
    if duplicate:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
user_scan()
```

### QUESTION 6: Agriculture – Premium Crop Price Filter

```python
def crop_filter():
    rate_list = list(map(int, input("Enter crop rates: ").split()))
    premium_list = []
    regular_list = []
    for rate in rate_list:
        if rate > 2000:
            premium_list.append(rate)
        else:
            regular_list.append(rate)
    print("Premium Crops:", premium_list)
    print("Regular Crops:", regular_list)
crop_filter()
```

### QUESTION 9: E-Learning – Student Attendance Eligibility System

```python
def attendance_report():
    record = list(map(int, input("Enter attendance record(with spaces): ").split()))
    attend_count = 0
    day_count = 0
    for day in record:
        day_count += 1
        if day == 1:
            attend_count += 1
    percent = (attend_count / day_count) * 100
    print("Attendance Percentage:", percent)
    if percent < 75:
        print("Exam Eligibility: Not Eligible")
    else:
        print("Exam Eligibility: Eligible")
attendance_report()
```

### QUESTION 10: Smart Electricity Bill Analyzer

```python
def power_bill():
    used_units = int(input("Enter units consumed: "))
    amount = 0
    if used_units <= 100:
        amount = used_units * 3
    elif used_units <= 200:
        first = 100 * 3
        second = (used_units - 100) * 5
        amount = first + second
    else:
        first = 100 * 3
        second = 100 * 5
        third = (used_units - 200) * 7
        amount = first + second + third
    if amount < 500:
        level = "Low Usage"
    elif amount <= 1500:
        level = "Moderate Usage"
    else:
        level = "High Usage"
    print("Total Bill:", amount)
    print("Usage Status:", level)
power_bill()
```
