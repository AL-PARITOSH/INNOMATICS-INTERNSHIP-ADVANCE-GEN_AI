### QUESTION 1: Employee Performance Bonus Eligibility

```python
x = int(input("Enter number of employees: "))
y = {}
for i in range(x):
    a = input("Enter name: ")
    b = int(input("Enter score: "))
    y[a] = b
c = max(y.values())
z = []
for i in y:
    if y[i] == c:
        z.append(i)
print("Top Performers Eligible for Bonus:")
print(", ".join(z), "(Score:", c, ")")
```

### QUESTION 2: Search Query Keyword Analysis

```python
x = input("Enter search query: ").lower()
a = ".,!?;:"
x = "".join(map(lambda c: "" if c in a else c, x))
y = x.split()
z = {}
for i in y:
    z[i] = z.get(i, 0) + 1
b = {}
for i in z:
    if z[i] > 1:
        b[i] = z[i]
print(b)
```

### QUESTION 3: Sensor Data Validation

```python
x = input("Enter sensor readings: ")
y = list(map(int, x.split()))
z = []
for i in range(len(y)):
    if y[i] % 2 == 0:
        z.append((i, y[i]))
print("Valid Sensor Readings (Hour, Value):")
print(z)
```

### QUESTION 4: Email Domain Usage Analysis

```python
x = int(input("Enter number of emails: "))
y = []
for i in range(x):
    a = input("Enter email: ")
    y.append(a)
z = list(map(lambda e: e.split("@")[1], y))
a = {}
for i in z:
    a[i] = a.get(i, 0) + 1
b = len(z)
for i in a:
    c = (a[i] / b) * 100
    print(i, ":", round(c), "%")
```

### QUESTION 5: Sales Spike Detection

```python
x = input("Enter sales values: ")
y = list(map(int, x.split()))
a = sum(y) / len(y)
print("Average Sales:", a)
for i in range(len(y)):
    if y[i] > a * 1.3:
        print("Day", i + 1, ":", y[i])
```

### QUESTION 6: Duplicate User ID Detection

```python
x = int(input("Enter number of user IDs: "))
y = []
for i in range(x):
    a = input("Enter user ID: ")
    y.append(a)
z = {}
for i in y:
    z[i] = z.get(i, 0) + 1
for i in z:
    if z[i] > 1:
        print(i, "→", z[i], "times")
```
