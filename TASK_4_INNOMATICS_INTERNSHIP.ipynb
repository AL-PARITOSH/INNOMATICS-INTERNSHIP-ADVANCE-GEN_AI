### QUESTION 1: Mobile Recharge Validation System

```python
def recharge():
    plans = [199, 299, 399, 599]
    print("Available Plans:", plans)
    while True:
        x = int(input("Enter recharge amount: "))
        if x < 50:
            print("Recharge amount must be at least ₹50.")
            print("Please try again.\n")
            continue
        if x not in plans:
            print("Invalid Plan Selected.")
            print("Valid Plans are:", plans)
            print("Please try again.\n")
            continue
        print("Recharge Successful!")
        print("You have successfully recharged ₹", x)
        break
recharge()
```

### QUESTION 2: Inventory Reorder Alert System

```python
def check_stock(d):    
    for i in d:
        if d[i] < 15:
            print(i, "→ Reorder Alert (Stock:", d[i], ")")
        else:
            print(i, "→ Stock OK (Stock:", d[i], ")")
n = int(input("Enter number of products: "))
products = {}
for i in range(n):
    name = input("Enter product name: ")
    qty = int(input("Enter stock quantity: "))
    products[name] = qty
check_stock(products)
```

### QUESTION 3: Student Result Processing System

```python
def check_result(marks):    
    total = 0
    for i in marks:
        total = total + i
    avg = total / len(marks)
    if avg >= 50:
        return avg,"Pass"
    else:
        return avg,"Fail"
x = list(map(int, input("Enter student marks (space separated): ").split()))
avg,result = check_result(x)
print("Average:",avg)
print("Result:", result)
```

### QUESTION 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(km, peak):    
    base = 50
    rate = 12
    fare = base + (rate * km)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
while True:    
    d = float(input("Enter distance in km: "))
    p = input("Is it peak hour? (yes/no): ").lower()
    total = calculate_fare(d, p)
    print("Total Fare: ₹", total)
    choice = input("Do you want to calculate again? (yes/no): ").lower()
    if choice != "yes":
        print("Thank you!,Check Again")
        break
```

### QUESTION 5: Employee Attendance Eligibility Checker

```python
def check_attendance(att):    
    total_days = len(att)
    present_days = 0
    for i in att:
        if i == "P":
            present_days = present_days + 1
    percentage = (present_days / total_days) * 100
    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"
x = input("Enter attendance (P and A separated by space): ").upper().split()
per, result = check_attendance(x)
print("Attendance Percentage:", per, "%")
print("Status:", result)
```

### QUESTION 6: Password Strength Checker

```python
def check_password(p):
    if len(p) < 8:
        return "Weak Password: Minimum length should be 8"
    has_digit = False
    has_special = False
    for i in p:
        if i.isdigit():
            has_digit = True
        if i in "@#$":
            has_special = True
    if not has_digit and not has_special:
        return "Weak Password: Digit and special character (@#$) should be present"
    elif not has_digit:
        return "Weak Password: At least one digit should be present"
    elif not has_special:
        return "Weak Password: At least one special character (@#$) should be present"
    else:
        return "Strong Password"
x = input("Enter password: ")
result = check_password(x)
print("Password Status:", result)
```
