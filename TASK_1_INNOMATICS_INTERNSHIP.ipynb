```python
#Question 2:PASS/FAIL ANALYZER
```

```python
marks = list(map(int, input("Enter marks separated by space: ").split()))

p_c = 0
f_c = 0

for m in marks:
    if m >= 50:
        p_c += 1
    else:
        f_c += 1

print("\nFinal Result:")
print("Total Students:", len(marks))
print("Total Pass Students:", p_c)
print("Total Fail Students:", f_c)
```

```python
#Question 3:SIMPLE DATA CLEANER
```

```python
names = list(map(str, input("Enter names separated by space: ").split()))

clean = list(map(lambda x: x.strip().lower(), names))

print("\nCleaned Names Are:")
print(",".join(clean))
```

```python
#Question 4:MESSAGE LENGTH ANALYZER
```

```python
messages = input("Enter messages separated by comma: ").split(",")

print("\nMessage Analysis:")

for i in messages:
    msg = i.strip()
    length = len(msg)

    print(f"Message: '{msg}'")
    print("Length:", length)

    if length > 10:
        print("Status: Longer than 10 characters")
    else:
        print("Status: OK")

    print()
```

```python
#Question 5:ERROR MESSAGE DETECTOR
```

```python
logs = input("Enter log entries separated by space: ").split()

e_c = 0

for i in logs:
    if i == "ERROR":
        e_c += 1

print("\nTotal ERROR entries:", e_c)
```
